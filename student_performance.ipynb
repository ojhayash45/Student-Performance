# **Univariate Analysis (Single Variable)**

```python
plt.figure(figsize=(6,4))
sns.histplot(df['GPA'], bins=20, kde=True)
plt.title("Distribution of GPA")
plt.show()
```

Above visuals shows that approx 70% of the student have GPA of range 1.5 - 2.5

```python
sns.histplot(df['Age'], bins=10)
plt.title("Age Distribution")
plt.show()
```

Above visuals shows that most of the students are near to age to 18.

## **Correlation** **Analysis**

```python
corr = df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

The correlation matrix shows the relationships between different student attributes and academic performance. It indicates that Absences have a strong negative correlation with GPA, while StudyTimeWeekly and ParentalSupport show a slight positive relationship with GPA, suggesting these factors influence student performance.

# **Bivariate Analysis**

```python
sns.scatterplot(x='StudyTimeWeekly', y='GPA', data=df)
plt.title("Study Time vs GPA")
plt.show()
```

The scatter plot shows that students who spend more weekly study time generally tend to achieve slightly higher GPAs, although the relationship is not very strong.

```python
sns.scatterplot(x='Absences', y='GPA', data=df)
plt.title("Absences vs GPA")
plt.show()
```

The scatter plot shows a strong negative relationship between absences and GPA, indicating that students with more absences generally have lower GPAs.

```python
sns.boxplot(x='Gender', y='GPA', data=df)
plt.title("Gender vs GPA")
plt.show()
```

The boxplot shows that GPA distribution is quite similar across both genders, indicating that gender does not significantly impact academic performance in this dataset.
